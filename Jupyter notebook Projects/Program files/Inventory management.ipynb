```python
import time

fd = open("Inventory.txt", "r")
products = fd.read().split("\n")
fd.close()

print("Welcome to the Store ğŸ›’")
print("-" * 40)

ui_user_name = input("Enter your name: ğŸ‘¤ ").strip()
print(f"Name      :  {ui_user_name}")

ui_phone = input("Enter your phone number : ğŸ“ ").strip()
print(f"Phone no. :  {ui_phone}")

ui_mail = input("Enter your E-mail id : ğŸ“§ ").strip()
print(f"E-mail    :  {ui_mail}")

print("-" * 40)
print("The Products we are having : ğŸ“¦")
print("Product id    Product name    Price ğŸ’°")

for product in products:
    if product.strip():
        prod_details_interfrace = product.split(",")
        print(f"{prod_details_interfrace[0]}. {prod_details_interfrace[1]},{prod_details_interfrace[2]}")

ui_prod_id = input("Enter the Product Id to search: ğŸ” ").strip()
ui_prod_qty = input("Enter the Quantity needed: ğŸ”¢ ").strip()

updated_prod_lst = []

for product in products:
    product = product.strip()
    if not product:
        continue
    prod_details = product.split(",")
    if prod_details[0] == ui_prod_id:
        if int(ui_prod_qty) <= int(prod_details[3]):
            print('-' * 40)
            print(f"The name of the product     : {prod_details[1]}")
            print(f"The price of the product    : {prod_details[2]}")
            print(f"The Quantity of the product : {ui_prod_qty}")
            print('-' * 40)
            print(f"Billing amount              : {int(prod_details[2]) * int(ui_prod_qty)} ğŸ’µ")
            print('-' * 40)

            fd = open("Transaction History.txt", "a")
            prod_details[3] = str(int(prod_details[3]) - int(ui_prod_qty))
            sales_record = f"{ui_user_name}, {ui_phone}, {ui_mail}, {prod_details[1]}, {ui_prod_qty}, {str(int(prod_details[2]) * int(ui_prod_qty))}, {time.strftime('%I:%M %p / %d-%m-%Y  ')}\n"
            fd.write(sales_record)
            print("Transaction Recorded Successfully âœ…")
            fd.close()
        else:
            print('-' * 40)
            print("Insufficient Quantity in stock âŒ")
            print(f"We only have {prod_details[3]} items in stock")
            print("Do you want to purchase the available quantity? (yes/no)")
            re_order = input("Press (yes/no) : ").strip()

            if re_order.lower() == "yes":
                print('-' * 40)
                print(f"The name of the product     : {prod_details[1]}")
                print(f"The price of the product    : {prod_details[2]}")
                print(f"The Quantity of the product : {prod_details[3]}")
                print('-' * 40)
                print(f"Billing amount              : {int(prod_details[2]) * int(prod_details[3])} ğŸ’µ")
                print('-' * 40)
                # Record transaction
                fd = open("Transaction History.txt", "a")
                qty_purchased = prod_details[3]
                sales_record = f"{ui_user_name}, {ui_phone}, {ui_mail}, {prod_details[1]}, {prod_details[3]}, {str(int(prod_details[2]) * int(ui_prod_qty))}, {time.strftime('%I:%M %p / %d-%m-%Y  ')}\n"
                fd.write(sales_record)
                print("Transaction Recorded Successfully âœ…")
                fd.close()
                prod_details[3] = '0'
            else:
                print("Thank you for visiting our store ğŸ™")

    updated_prod_lst.append(prod_details)

lst = []

for i in updated_prod_lst:
    prod = f"{i[0]},{i[1]},{i[2]},{i[3]}\n"
    lst.append(prod)

if lst:
    lst[-1] = lst[-1][:-1]

fd = open('Inventory.txt', 'w')
for i in lst:
    fd.write(i)

print("-" * 40)
print("Inventory updated successfully ğŸ”„")
fd.close()
```
